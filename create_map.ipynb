```python
import folium
from geopy.geocoders import Nominatim
import time

# Címek
addresses = [
    "Budapest, Csobánc utca 5",
    "Budapest, Csobánc utca 1",
    "Budapest, Mátyás tér 4",
    "Budapest, Mátyás tér 15",
    "Budapest, Vajda Péter utca 37-39",
    "Budapest, Vajda Péter utca 25-31"
]

# Geocoder
geolocator = Nominatim(user_agent="my_map_app")

locations = []

for addr in addresses:
    location = geolocator.geocode(addr)
    if location:
        locations.append((addr, location.latitude, location.longitude))
    else:
        print(f"Nem található: {addr}")
    time.sleep(1)  # fontos: ne spameld az API-t

# Térkép középpont
map_center = [47.485, 19.095]

m = folium.Map(location=map_center, zoom_start=14)

# Pontok + körök
for name, lat, lon in locations:
    folium.Marker(
        location=[lat, lon],
        popup=name
    ).add_to(m)

    folium.Circle(
        radius=150,
        location=[lat, lon],
        color="red",
        fill=True,
        fill_opacity=0.2
    ).add_to(m)

# Mentés
m.save("terkep.html")

print("Kész! terkep.html létrehozva.")

```
